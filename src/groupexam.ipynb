```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
import string
import nltk
import sklearn 
import mglearn
import joblib
import os
# import tensorflow as tf
# import keras 
```

```python
#Load the reviews dataset
df = pd.read_csv('../reviews.csv')

#Display basic information about the dataset
print("Dataset Shape:", df.shape)
print("\n first few rows of the dataset:")
print(df.head())
print("\n Datatype:")
print(df.dtypes)
print("\n Missing values:")
print(df.isnull().sum())
print("\n Relevant statistics:") # TODO: bad wording
print(df.describe()) # TODO: format to 2 decimals

```

```python
#Examine the target variable (score distribution)
print("Score Distribution:")
print(df['rating'].value_counts().sort_index())
print("\n Score Distribution (%):")
print((df['rating'].value_counts(normalize=True).sort_index() * 100).round(2))

#Visualize the score distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
df['rating'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color='blue')
axes[0].set_title('Score Distribution (Count)')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Count')

(df['rating'].value_counts(normalize=True).sort_index().plot(kind='bar', ax=axes[1], color='red'))
axes[1].set_title('Score Distribution (%)')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('Percentage')

plt.tight_layout()
plt.show() # TODO: remove dtype: int64 ???

print("\nClass imbalance analysis") # TODO: better wording
print("This distribution shows that the dataset is imbalanced, with a higher number of reviews having a rating of 5 compared to other ratings. This imbalance may affect the performance of machine learning models trained on this data, as they may be biased towards predicting the majority class (rating of 5). It is important to consider techniques for handling class imbalance, such as resampling methods or using evaluation metrics that account for imbalance, when developing models with this dataset.")

```
